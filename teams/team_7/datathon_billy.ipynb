```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, GlobalAveragePooling2D, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from keras.applications.densenet import DenseNet121
```

```python
joined_dat = joined_dat.rename(columns={'Sex':'Male'})
x = joined_dat.iloc[:,:6]
y = joined_dat.iloc[:,6:]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
# Set the image dimensions
img_width, img_height = 128, 128

# Load and preprocess images
def load_and_preprocess_image(image_path):
    img = load_img(image_path, target_size=(img_width, img_height))
    img_array = img_to_array(img) / 255.0  # Normalize pixel values
    return img_array

# Load and preprocess all images
images_train = [load_and_preprocess_image(path) for path in X_train['ImagePath']]

# Load and preprocess all images
images_test = [load_and_preprocess_image(path) for path in X_test['ImagePath']]
```

```python
# joined_dat.to_csv('joined_dat_images.csv', index=False)
```

```python
# joined_dat = pd.read_csv('joined_dat_images.csv')
```

```python
model_d=DenseNet121(weights='imagenet',include_top=False, input_shape=(128, 128, 3)) 

x=model_d.output

x= GlobalAveragePooling2D()(x)
x= BatchNormalization()(x)
x= Dropout(0.5)(x)
x= Dense(1024,activation='relu')(x) 
x= Dense(512,activation='relu')(x) 
x= BatchNormalization()(x)
x= Dropout(0.5)(x)

preds=Dense(1,activation='sigmoid')(x) #FC-layer

model=Model(inputs=model_d.input,outputs=preds)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
num_epochs = 5  # Number of training epochs

for target in joined_dat.columns[6:]:
    # Convert lists to numpy arrays
    labels = np.array(y_train[target])
    images_train_array = np.array(images_train)
    images_test_array = np.array(images_test)
    
    # Train the model
    model.fit(
        images_train_array,
        labels,
        epochs=num_epochs
    )

    new_column_name = target + '_bl_prob'
    y_test[new_column_name] = model.predict(images_test_array)
    X_train[new_column_name] = model.predict(images_train_array)
    print(target + ' done')
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc

# train auc
print(roc_auc_score(y_train['No Finding'], X_train['No Finding_bl_prob']))
# test auc
print(roc_auc_score(y_test['No Finding'], y_test['No Finding_bl_prob']))
```

```python
from sklearn.ensemble import RandomForestClassifier

# [Age, Male, White, Black, Other]
feat_list = ['No Finding_bl_prob']
for feat in X_train.columns[:5]:
    feat_list.append(feat)
    new_X_train = X_train[feat_list] # 5000
    new_X_test = pd.concat([y_test['No Finding_bl_prob'], X_test[feat]], axis=1) # 2000
    model = RandomForestClassifier(n_estimators=100).fit(new_X_train, y_train['No Finding'])
    temp_name = 'NN_' + feat
    y_train[temp_name] = model.predict_proba(new_X_train)[:,1]
    y_test[temp_name] = model.predict_proba(new_X_test)[:,1]
    print(roc_auc_score(y_train['No Finding'], y_train[temp_name]))
    print(roc_auc_score(y_test['No Finding'], y_test[temp_name]))
    print()

    del feat_list[-1]
```
